```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
def generate_normal_time_series(num, minl=50, maxl=1000):
    """Purpose: generate random series
    with changepoints,
    taken from:
    http://nbviewer.jupyter.org/github/hildensia/bayesian_changepoint_detection/blob/master/Example%20Code.ipynb
    """
    data = np.array([], dtype=np.float64)
    partition = np.random.randint(minl, maxl, num)
    for p in partition:
        mean = np.random.randn()*10
        var = np.random.randn()*1
        if var < 0:
            var = var * -1
        tdata = np.random.normal(mean, var, p)
        data = np.concatenate((data, tdata))
    return data
```

```python
# generate data with a few changepoints
data = generate_normal_time_series(3, 50, 3500)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data, '#34495e')
plt.title('Time Series with Change Points')
plt.show()
```

** Get Changepoints **

```python
from ChangePointDetector import ChangePointDetector
```

```python
cp_detector = ChangePointDetector(data=data, num_changepoints=3)
```

```python
# get changepoint locations and predicted mean
locs, preds_locs = cp_detector.calculate_changepoint_locations()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data, '#34495e', label='Raw Data')
for j in locs:
    plt.plot(j, np.mean(data), 'ro')
plt.title('Time Series with Change Points')
plt.plot(preds_locs, 'r', label='Predicted Mean')
plt.legend(loc='lower right')
plt.show()
```
